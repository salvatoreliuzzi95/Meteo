# Data Gatherer v1.5

```python
import os
import sys
import pandas as pd  
from datetime import datetime
  
df = pd.DataFrame()  

DIR  = sys.path[0]
path = os.path.join(DIR,'Raw','378.txt')

start_line = 1068-1

f     = open(path, "r")
lines = f.readlines()
```

```python
print(lines)
```

```python
# Grafico del vento

#GV = lines[776] 
#GV = lines[779] 
#GV = lines[780] 
#GV = lines[783] 
#GV = lines[784]

GV = lines[start_line] 

GV = GV.replace("                        ","")
GV = GV.replace("},{","} {")
GV = GV.replace("{ x: ","")
GV = GV.replace("marker: { width:64, height:64, symbol: 'url(/img/imagechart/windDirection_N.png)'},","") 
GV = GV.replace("marker: { width:64, height:64, symbol: 'url(/img/imagechart/windDirection_NNE.png)'},","") 
GV = GV.replace("marker: { width:64, height:64, symbol: 'url(/img/imagechart/windDirection_NE.png)'},","") 
GV = GV.replace("marker: { width:64, height:64, symbol: 'url(/img/imagechart/windDirection_ENE.png)'},","") 
GV = GV.replace("marker: { width:64, height:64, symbol: 'url(/img/imagechart/windDirection_E.png)'},","") 
GV = GV.replace("marker: { width:64, height:64, symbol: 'url(/img/imagechart/windDirection_SE.png)'},","") 
GV = GV.replace("marker: { width:64, height:64, symbol: 'url(/img/imagechart/windDirection_ESE.png)'},","") 
GV = GV.replace("marker: { width:64, height:64, symbol: 'url(/img/imagechart/windDirection_SSE.png)'},","") 
GV = GV.replace("marker: { width:64, height:64, symbol: 'url(/img/imagechart/windDirection_S.png)'},","") 
GV = GV.replace("marker: { width:64, height:64, symbol: 'url(/img/imagechart/windDirection_SSW.png)'},","") 
GV = GV.replace("marker: { width:64, height:64, symbol: 'url(/img/imagechart/windDirection_SW.png)'},","") 
GV = GV.replace("marker: { width:64, height:64, symbol: 'url(/img/imagechart/windDirection_WSW.png)'},","") 
GV = GV.replace("marker: { width:64, height:64, symbol: 'url(/img/imagechart/windDirection_W.png)'},","") 
GV = GV.replace("marker: { width:64, height:64, symbol: 'url(/img/imagechart/windDirection_WNW.png)'},","") 
GV = GV.replace("marker: { width:64, height:64, symbol: 'url(/img/imagechart/windDirection_NW.png)'},","") 
GV = GV.replace("marker: { width:64, height:64, symbol: 'url(/img/imagechart/windDirection_NNW.png)'},","") 
GV = GV.replace("Date.UTC(","") 
GV = GV.replace(")","") 

GV = GV.replace("01-1","1") 
GV = GV.replace("02-1","2") 
GV = GV.replace("03-1","3") 
GV = GV.replace("04-1","4") 
GV = GV.replace("05-1","5") 
GV = GV.replace("06-1","6") 
GV = GV.replace("07-1","7") 
GV = GV.replace("08-1","8") 
GV = GV.replace("09-1","9") 
GV = GV.replace("10-1","10") 
GV = GV.replace("11-1","11") 
GV = GV.replace("12-1","12") 

GV = GV.replace(" y: ","") 
GV = GV.replace(" 'cardinalPoint': ","") 
GV = GV.replace(" 'windDirection':","") 
GV = GV.replace("'","") 
GV = GV.replace(" ","") 
GV = GV.replace(",\n","") 
  
GV = list(GV.split("}"))

df["data"]                      = 0
df["velocità vento [km/h]"]     = 0
df["cardinalità vento"]         = 0
df["direzione vento [°]"]       = 0

for row in range(0,len(GV)-1):
    df.loc[row,"data"]                  = str(GV[row].split(",")[0])+"-"+\
                                          str(GV[row].split(",")[1])+"-"+\
                                          str(GV[row].split(",")[2])+" "+\
                                          str(GV[row].split(",")[3])+":"+\
                                          str(GV[row].split(",")[4])

    df.loc[row,"velocità vento [km/h]"] = GV[row].split(",")[5]
    df.loc[row,"cardinalità vento"]     = GV[row].split(",")[6]
    df.loc[row,"direzione vento [°]"]   = GV[row].split(",")[7]

df['data'] = pd.to_datetime(df['data'])

df.tail()
```
