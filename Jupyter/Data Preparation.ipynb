# Data Preparation v1.2

```python
import os
import sys
import pandas as pd  
import datetime
from datetime import datetime

DIR = sys.path[0]
```

```python
csv      = 36
csv_name = "df"+str(csv)+".csv"
test     = pd.read_csv(os.path.join(DIR,'Data',csv_name),sep = ",").drop(columns="Unnamed: 0");
dfs      = [test]
```

```python
dataset = pd.concat(dfs).drop_duplicates().reset_index(drop=True)
```

```python
dataset.to_csv(r'testset.csv')
```

```python

temp_dataset = dataset

for row in range (0,len(dataset)):
    if(row>0):
        data_a  = pd.to_datetime(dataset.loc[row-1,"data"])
        data_b  = pd.to_datetime(dataset.loc[row,"data"])
        dt_pace = pd.Timedelta("0 days 00:05:00")
        gap     = int((data_b-data_a)/("0 days 00:05:00"))
        
        vv_a    = dataset.loc[row-1,"velocità vento [km/h]"]
        vv_b    = dataset.loc[row  ,"velocità vento [km/h]"]
        vv_pace = abs((vv_b-vv_a))/(gap+0.000000000001)
        
        dv_a    = dataset.loc[row-1,"direzione vento [°]"]
        dv_b    = dataset.loc[row  ,"direzione vento [°]"]
        dv_pace = abs((dv_b-dv_a))/(gap+0.000000000001)
        
        pc_a    = dataset.loc[row-1,"pioggia cumulata [mm]"]
        pc_b    = dataset.loc[row  ,"pioggia cumulata [mm]"]
        pc_pace = abs((pc_b-pc_a))/(gap+0.000000000001)
                
        pi_a    = dataset.loc[row-1,"pioggia istantanea [mm]"]
        pi_b    = dataset.loc[row  ,"pioggia istantanea [mm]"]
        pi_pace = abs((pi_b-pi_a))/(gap+0.000000000001)
                
        p_a     = dataset.loc[row-1,"pressione [hPa]"]
        p_b     = dataset.loc[row  ,"pressione [hPa]"]
        p_pace  = abs((p_b-p_a))/(gap+0.000000000001)  
        
        tr_a    = dataset.loc[row-1,"temperatura di rugiada [°C]"]
        tr_b    = dataset.loc[row  ,"temperatura di rugiada [°C]"]
        tr_pace = abs((tr_b-tr_a))/(gap+0.000000000001) 
        
        ur_a    = dataset.loc[row-1,"umidità relativa [%]"]
        ur_b    = dataset.loc[row  ,"umidità relativa [%]"]
        ur_pace = abs((ur_b-ur_a))/(gap+0.000000000001)
        
        t_a     = dataset.loc[row-1,"temperatura [°C]"]
        t_b     = dataset.loc[row  ,"temperatura [°C]"]
        t_pace  = abs((t_b-t_a))/(gap+0.000000000001)
        
    
        if(gap>1):
            print("GAP FOUND ***********************************************************************************************")
            print("Data A:          "+str(data_a))
            print("Data B:          "+str(data_b))
            print("Missing Records: "+str(gap-1))
            
            for i in range (1,gap): 
                data_c = data_a       + i*dt_pace
                
                vv_c = 0
                dv_c = 0
                pc_c = 0
                pi_c = 0
                p_c  = 0
                tr_c = 0
                ur_c = 0
                t_c  = 0
                
                if(vv_a<=vv_b): vv_c = round(vv_a + i*vv_pace,3)
                if(vv_a> vv_b): vv_c = round(vv_a - i*vv_pace,3)
            
                if(dv_a<=dv_b): dv_c = round(dv_a + i*dv_pace,3)
                if(dv_a> dv_b): dv_c = round(dv_a - i*dv_pace,3)
                
                if(pc_a<=pc_b): pc_c = round(pc_a + i*pc_pace,3)
                if(pc_a> pc_b): pc_c = round(pc_a - i*pc_pace,3)
                    
                if(pi_a<=pi_b): pi_c = round(pi_a + i*pi_pace,3)
                if(pi_a> pi_b): pi_c = round(pi_a - i*pi_pace,3)
                    
                if(p_a<=p_b):   p_c  = round(p_a  + i*p_pace ,3)
                if(p_a> p_b):   p_c  = round(p_a  - i*p_pace ,3)       
                    
                if(tr_a<=tr_b): tr_c = round(tr_a + i*tr_pace,3)
                if(tr_a> tr_b): tr_c = round(tr_a - i*tr_pace,3)                     
                
                if(ur_a<=ur_b): ur_c = round(ur_a + i*ur_pace,3)
                if(ur_a> ur_b): ur_c = round(ur_a - i*ur_pace,3) 
                
                if(t_a<=t_b):   t_c  = round(t_a  + i*t_pace ,3)
                if(t_a> t_b):   t_c  = round(t_a  - i*t_pace ,3)
                
                vdc = "none"
                
                if (dv_c <= 11.25):                    vdc = 'N'       
                if (dv_c >= 348.75):                   vdc = 'N'       
                if (dv_c > 11.25  and dv_c <= 33.75):  vdc = 'NNE'
                if (dv_c > 33.75  and dv_c <= 56.25):  vdc = 'NE'
                if (dv_c > 56.25  and dv_c <= 78.75):  vdc = 'ENE'
                if (dv_c > 78.75  and dv_c <= 101.25): vdc = 'E'
                if (dv_c > 101.25 and dv_c <= 123.75): vdc = 'ESE'
                if (dv_c > 123.75 and dv_c <= 146.25): vdc = 'SE'
                if (dv_c > 146.25 and dv_c <= 168.75): vdc = 'SSE'
                if (dv_c > 168.75 and dv_c <= 191.25): vdc = 'S'
                if (dv_c > 191.25 and dv_c <= 213.75): vdc = 'SSW'
                if (dv_c > 213.75 and dv_c <= 236.25): vdc = 'SW'
                if (dv_c > 236.25 and dv_c <= 258.75): vdc = 'WSW'
                if (dv_c > 258.75 and dv_c <= 281.25): vdc = 'W'
                if (dv_c > 281.25 and dv_c <= 303.75): vdc = 'WNW'
                if (dv_c > 303.75 and dv_c <= 326.25): vdc = 'NW'
                if (dv_c > 326.75 and dv_c <= 348.75): vdc = 'NNW'
                if (dv_c == 0     ): vdc = 'N'
                if (dv_c == 22.5  ): vdc = 'NNE'
                if (dv_c == 45    ): vdc = 'NE'
                if (dv_c == 67.5  ): vdc = 'ENE'
                if (dv_c == 90    ): vdc = 'E'
                if (dv_c == 112.5 ): vdc = 'ESE'
                if (dv_c == 135   ): vdc = 'SE'
                if (dv_c == 157.5 ): vdc = 'SSE'
                if (dv_c == 180   ): vdc = 'S'
                if (dv_c == 202.5 ): vdc = 'SSW'
                if (dv_c == 225   ): vdc = 'SW'
                if (dv_c == 247.5 ): vdc = 'WSW'
                if (dv_c == 270   ): vdc = 'W'
                if (dv_c == 292.5 ): vdc = 'WNW'
                if (dv_c == 315   ): vdc = 'NW'
                if (dv_c == 337.5 ): vdc = 'NNW'
                if (dv_c == 360   ): vdc = 'N'
                                        
                string = str(data_c)+" "\
                        +str(vv_c)+" "\
                        +str(dv_c)+" "\
                        +str(pc_c)+" "\
                        +str(pi_c)+" "\
                        +str(p_c) +" "\
                        +str(tr_c)+" "\
                        +str(ur_c)+" "\
                        +str(t_c) +" "\
                        +vdc
                
                dataset.loc[len(dataset)+1,"data"]                        = data_c
                dataset.loc[len(dataset)  ,"velocità vento [km/h]"]       = vv_c
                dataset.loc[len(dataset)  ,"cardinalità vento"]           = vdc
                dataset.loc[len(dataset)  ,"direzione vento [°]"]         = dv_c
                dataset.loc[len(dataset)  ,"pioggia cumulata [mm]"]       = pc_c
                dataset.loc[len(dataset)  ,"pioggia istantanea [mm]"]     = pi_c
                dataset.loc[len(dataset)  ,"pressione [hPa]"]             = p_c
                dataset.loc[len(dataset)  ,"temperatura di rugiada [°C]"] = tr_c
                dataset.loc[len(dataset)  ,"umidità relativa [%]"]        = ur_c
                dataset.loc[len(dataset)  ,"temperatura [°C]"]            = t_c

                print(string)
                
# note: forse c'è un bug con i missing records

```

```python
dataset['data'] = pd.to_datetime(dataset['data'])
dataset = dataset.sort_values(by=['data'])
dataset.reset_index(drop=True, inplace=True)

dataset.to_csv(r'dataset.csv')
```

```python
hours_earlier   = 3
time_window_h   = 3
init_steps_back = int((hours_earlier)*60/5 - 1)
steps_back      = int((hours_earlier+time_window_h)*60/5)

#hours_earlier   = 0 
#time_window_h   = 2
#init_steps_back = 0
#steps_back      = int((hours_earlier+time_window_h)*60/5)

actual_sb       = steps_back - init_steps_back
```

```python
df = dataset

for i in range(0,actual_sb):
    
    print ("Creating (n-"+str(i+1+init_steps_back)+") features...")
    
    vv_sb  =  "velocità vento [km/h] (n-"       +str(i+1+init_steps_back)+")"
    cv_sb  =  "cardinalità vento (n-"           +str(i+1+init_steps_back)+")"
    dv_sb  =  "direzione vento [°] (n-"         +str(i+1+init_steps_back)+")"
    pc_sb  =  "pioggia cumulata [mm] (n-"       +str(i+1+init_steps_back)+")"
    pi_sb  =  "pioggia istantanea [mm] (n-"     +str(i+1+init_steps_back)+")"
    p_sb   =  "pressione [hPa] (n-"             +str(i+1+init_steps_back)+")"
    tr_sb  =  "temperatura di rugiada [°C] (n-" +str(i+1+init_steps_back)+")"    
    ur_sb  =  "umidità relativa [%] (n-"        +str(i+1+init_steps_back)+")"
    t_sb   =  "temperatura [°C] (n-"            +str(i+1+init_steps_back)+")"

    df[vv_sb] = 0
    df[cv_sb] = 0
    df[dv_sb] = 0
    df[pc_sb] = 0
    df[pi_sb] = 0
    df[p_sb]  = 0
    df[tr_sb] = 0
    df[ur_sb] = 0
    df[t_sb]  = 0


    for row in range(0, len(df)):
        if(row>i):
              
            df.loc[row,vv_sb]  =  df.loc[row-(i+1),"velocità vento [km/h]"]
            df.loc[row,cv_sb]  =  df.loc[row-(i+1),"cardinalità vento"]
            df.loc[row,dv_sb]  =  df.loc[row-(i+1),"direzione vento [°]"]
            df.loc[row,pc_sb]  =  df.loc[row-(i+1),"pioggia cumulata [mm]"]
            df.loc[row,pi_sb]  =  df.loc[row-(i+1),"pioggia istantanea [mm]"]
            df.loc[row, p_sb]  =  df.loc[row-(i+1),"pressione [hPa]"]
            df.loc[row,tr_sb]  =  df.loc[row-(i+1),"temperatura di rugiada [°C]"]
            df.loc[row,ur_sb]  =  df.loc[row-(i+1),"umidità relativa [%]"]
            df.loc[row, t_sb]  =  df.loc[row-(i+1),"temperatura [°C]"]


df.tail()
```

```python
temp_dataframe  = df;
indexesTD = []
for row in range(0,actual_sb): #-1?
    indexesTD.append(row)
temp_dataframe = temp_dataframe.drop(indexesTD) 
temp_dataframe.reset_index(drop=True, inplace=True)
print(str(len(indexesTD))+" records deleted, "+str(len(temp_dataframe))+" left")
df = temp_dataframe;

df.head()

dataset = df
```

```python
import datetime

dataset['ora']=0

for row in range(0,len(dataset)):
    temp = int(dataset.loc[row,'data'].hour)*60 + int(dataset.loc[row,'data'].minute) + hours_earlier*60
    if (temp >= 24*60): 
        temp = temp - 24*60
    temp = datetime.timedelta(minutes=temp)
    dataset.loc[row,'ora'] = temp
    
dataset.head()
```
